```python
import tensorflow as tf
import numpy as np
```

```python
data=np.random.normal(10,1,100) #100 no with mean 10 and sd 1
alpha=tf.constant(0.05)
current=tf.placeholder(tf.float32)
prev_avg=tf.Variable(0.)
update_avg=alpha*current+(1-alpha)*prev_avg

init=tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(len(data)):
        curr_avg=sess.run(update_avg,feed_dict={current:data[i]})
        sess.run(tf.assign(prev_avg,curr_avg))
        print(data[i],curr_avg)
```
