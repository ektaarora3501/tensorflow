```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X_val=pd.read_csv("/home/ekta/python_codes/ml/tensorflow/kc_house_data.csv", usecols=[5],nrows=50,header=None,skiprows=[0])
Y_val=pd.read_csv("/home/ekta/python_codes/ml/tensorflow/kc_house_data.csv", usecols=[2],nrows=50,header=None,skiprows=[0])
x_train=np.float32(X_val)
y_train=np.float32(Y_val)
#print(x_train,y_train)
plt.scatter(x_train,y_train)
plt.show()
```

```python
learning_rate=0.01

X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
print(X)

def model(X,w):
    return tf.multiply(X,w)

w=tf.Variable(0.0,name="weight")

y_model=model(X,w)
cost=tf.square(Y-y_model)

train_op=tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

sess=tf.Session()
init=tf.global_variables_initializer()
sess.run(init)

for i in range(0,50):
    for (x,y) in zip(x_train,y_train):
        sess.run(train_op,feed_dict={X:x,Y:y})

w_val=sess.run(w)
print(w_val)

sess.close()
plt.scatter(x_train,y_train)
y_learned=x_train*w_val
plt.plot(x_train,y_learned,'r')
plt.show()
        
```
